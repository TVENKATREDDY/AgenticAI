```python
model=LiteLLMModel(
    model_id="gemini/gemini-1.5-flash",
    api_key=os.getenv("GOOGLE_API_KEY")

)
```

```python
ml_code_agent=CodeAgent(
    tools=[DuckDuckGoSearchTool()],
    additional_authorized_imports=['pandas','numpy','sklearn','json'],
    model=model
)
```

```python
def generate_ml_code(prompt):
    full_prompt="""
        You are an AI assistant helping with machine learning code generation.
        complte below 9 steps in structured manner.
        1. **Load the diabatic dataset** using sklearn
        2. **split the data** (80% train and 20% test).
        3. **Train a linear regression model**
        4. ** Find the best three feature combinations**.(Lowest RMSE for the model).
        5. ** Find the R2 & adjusted R2 ** for the model.
        6. ** Apply feature engineering** if required.
        7. **List the top 5 models based on RMSE**.
        8. ** Output the results in structured format**.
        9. ** Plot the result**.
        """
    try:
        response=ml_code_agent.run(full_prompt)
        return response
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
infa=gr.Interface(
    fn=generate_ml_code,
    inputs="text",
    outputs="text",
    title='ML Code generator using (Gemini 1.5 flash)',
    description="Enter a Machine Learning coding task and AI will generate python code",
)
```

```python
infa.launch()
```
